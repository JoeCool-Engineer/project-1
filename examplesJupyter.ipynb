### Search Result for "plot eigenvectors on pairplot python"
To plot eigenvectors on a pairplot in Python, it's necessary to first perform Principal Component Analysis (PCA) to obtain the eigenvectors. Then, these eigenvectors can be visualized as vectors originating from the mean of the data in the pairplot. Here's how it can be done:
In this code:
* **Data Preparation:** Sample data is created, standardized, and PCA is applied to reduce dimensionality to 2 components.
* **Eigenvector Calculation:** Eigenvectors are extracted from the PCA results.
* **Pairplot Creation:** A pairplot is generated using the PCA-transformed data.
* **Eigenvector Overlay:** Eigenvectors are plotted as arrows originating from the mean of the original data, scaled to fit the plot.
* **Visualization:** The plot is displayed with eigenvectors overlaid.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Sample data (replace with your actual data)
data = np.random.rand(100, 3)
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3'])

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Perform PCA
pca = PCA(n_components=2)  # Reduce to 2 components for visualization
pca.fit(scaled_data)
pca_data = pca.transform(scaled_data)

# Eigenvectors
eigenvectors = pca.components_

# Mean of the original data
mean_data = np.mean(scaled_data, axis=0)

# Create a Pandas DataFrame for the PCA data
pca_df = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2'])

# Plot the pairplot
plt.figure(figsize=(8, 8))
sns.pairplot(pca_df)

# Overlay eigenvectors
for i, v in enumerate(eigenvectors):
    plt.gca().arrow(0, 0, v[0], v[1], head_width=0.1, head_length=0.1, fc='red', ec='red', label=f'Eigenvector {i+1}')

plt.show()
```
